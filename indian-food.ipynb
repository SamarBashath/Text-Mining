```r
food$Text = gsub("&amp", "", food$Text)
```

```r
food$Text = gsub("(RT|via)((?:\\b\\W*@\\w+)+)", "", food$Text)
food$Text = gsub("@\\w+", "", food$Text)
food$Text = gsub("[[:punct:]]", "", food$Text)
food$Text = gsub("[[:digit:]]", "", food$Text)
food$Text = gsub("http\\w+", "", food$Text)
food$Text = gsub("[ \t]{2,}", "", food$Text)
food$Text = gsub("^\\s+|\\s+$", "", food$Text)
food$Text <- iconv(food$Text, "UTF-8", "ASCII", sub="")
food$Text <- tolower(food$Text)
```

```r
my_stop_words <- tibble(   word = c( "fat","shit","fuck","know","get","will","can","try","food","foodies","delicious","eat","yummy","yum","tasty","home","day","just","recipe","like","good","work","going","dont","love","make","one","tomorrow","time","new","lunch","foodie","ufb","sleep","made", "knowing","breakfast","perfect","looks","foodporn","now","bts","help","great","clip","idfyum","ufacbts","idk","need","itstimeforbtsidyumpia","Ldfyumplay", "ufae","today","dinner","special","comfortfood","want","episode","idyum","Itstimeforbtsldfyumpla","ldfyumplay","best","happy","didn't","loss","really","eating","lol","stop","come","recipes","improve","something","aid","miss","sad","break","fucking","false","hamper","slowly","win","comfort","super","amazing","twist","shake","pan","free","rich","top","classic","wrong","lazy","dick","damn","stall", "stalls","fans" ),   lexicon ="twitter" )
```

```r
new_stop_list <- stop_words %>% 
bind_rows(my_stop_words)
```

```r
clean <- food %>%
anti_join(new_stop_list)
```

```r
food_sentiment1 <- clean %>%   
inner_join(get_sentiments("bing")) %>%   
count(word, sentiment)
```

```r
topwords <- food_sentiment1 %>%   
group_by(sentiment) %>%   
top_n(15) %>%   
ungroup() %>%   
mutate(word = reorder(word, n))
```

```r
ggplot(topwords, aes(word, n, fill= sentiment))+      
geom_col(show.legend = FALSE) +       
facet_wrap(~sentiment, scales = "free") + 
labs(title= "Food word Trend Tweets") + coord_flip()
```
